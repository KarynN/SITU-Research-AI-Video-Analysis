```python
import pathlib
import textwrap

import google.generativeai as genai

from IPython.display import display
from IPython.display import Markdown


def to_markdown(text):
  text = text.replace('â€¢', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

```python
# Or use `os.getenv('GOOGLE_API_KEY')` to fetch an environment variable.
GOOGLE_API_KEY='key'#userdata.get('GOOGLE_API_KEY')

genai.configure(api_key=GOOGLE_API_KEY)
```

```python
import PIL.Image

img = PIL.Image.open('/content/drive/MyDrive/output_frame4.jpg')
img
```

```python
model = genai.GenerativeModel('gemini-1.5-flash')

response = model.generate_content(img)

to_markdown(response.text)
```

```python
response = model.generate_content(["list all text and numbers in this surveillance image and explain what they mean", img])
to_markdown(response.text)
```
