# Save Videos as images (1 img/s)

```python
import pandas as pd
import numpy as np

csvpath = '/content/drive/MyDrive/Test_video/ugc_metadata.csv'

ugcvids = pd.read_csv(csvpath)

```

```python
import re
def convert_time_format(time_str):
    time_str = time_str.strip()

    # Match different time formats and convert them to timedelta
    if re.match(r'^\d+:\d+:\d+(\.\d+)?$', time_str):  # Match hh:mm:ss.sss
        h, m, s = time_str.split(':')
        return pd.Timedelta(hours=int(h), minutes=int(m), seconds=float(s))
    elif re.match(r'^\d+(\.\d+)?\s?s$', time_str):  # Match ss.sss or ss.sss s
        return pd.Timedelta(seconds=float(re.sub(r'\s?s$', '', time_str)))
    else:
        raise ValueError(f"Unknown format: {time_str}")


ugcvids['duration_format'] = ugcvids['Duration'].apply(convert_time_format)
ugcvids['duration_format'] = ugcvids['duration_format'].dt.total_seconds().apply(lambda x: pd.to_datetime(x, unit='s').strftime('%H:%M:%S.%f')[:-3])
```

```python
filtered = ugcvids[pd.to_timedelta(ugcvids['duration_format']).dt.total_seconds().between(60, 600)]

# Sample 100 entries or the maximum available if less than 100
sampled = filtered.sample(n=min(100, len(filtered)))
```

```python
sampled.head()
```

```python
import os
import cv2

base_folder = '/content/drive/MyDrive/Test_video/ugcsamples'
os.makedirs(base_folder, exist_ok=True)

# Process each sampled entry
for idx, (index, row) in enumerate(sampled.iterrows(), start=1):
    video_file = row['SourceFile']
    filename = os.path.splitext(row['FileName'])[0]
    output_folder = os.path.join(base_folder, filename)
    print(output_folder)
    os.makedirs(output_folder, exist_ok=True)

    # Open the video file
    cap = cv2.VideoCapture(video_file)
    fps = int(cap.get(cv2.CAP_PROP_FPS))  # Get frames per second
    frame_count = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Save one frame per second
        if frame_count % fps == 0:
            frame_number = frame_count // fps
            #frame_filename = os.path.join(output_folder, f'{output_folder}_frame_{frame_number:04d}.jpg')
            frame_filename = (f'{output_folder}/{filename}_frame_{frame_number:04d}.jpg')
            print(frame_filename)
            cv2.imwrite(frame_filename, frame)

        frame_count += 1

    cap.release()
    print(f"Entry {idx}: Processed {row['FileName']}")


```

```python
import pathlib
import textwrap

import google.generativeai as genai

from IPython.display import display
from IPython.display import Markdown


def to_markdown(text):
  text = text.replace('•', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

```python
model = genai.GenerativeModel('gemini-1.5-flash')
```

```python
from PIL import Image
image_folder = '/content/drive/MyDrive/Test_video/ugcsamples/DEF_000266301'
images = []

filenames = sorted(os.listdir(image_folder))

num_images = len(filenames)

interval = max(1, num_images // 10)

# Sample 10 equally spaced images
for i in range(0, num_images, interval):
    img_path = os.path.join(image_folder, filenames[i])
    img = Image.open(img_path)
    images.append(img)

if len(images) > 10:
    images = images[:10]

# Add the prompt to the list of images
prompt = "These are consecutive frames from a video. Describe this scene."
images.append(prompt)

response = model.generate_content(images)
to_markdown(response.text)
```

```python

base_folder = '/content/drive/MyDrive/Test_video/ugcsamples'
output_data = []

for folder_name in os.listdir(base_folder):
    image_folder = os.path.join(base_folder, folder_name)
    if os.path.isdir(image_folder):
        images = []
        image_paths = []

        filenames = sorted(os.listdir(image_folder))

        num_images = len(filenames)

        # Calculate the interval for equally spaced sampling
        interval = max(1, num_images // 10)

        # Sample 10 equally spaced images
        for i in range(0, num_images, interval):
            img_path = os.path.join(image_folder, filenames[i])
            img = Image.open(img_path)
            images.append(img)
            img_path = os.path.basename(img_path)
            image_paths.append(img_path)
        if len(images) > 10:
            images = images[:10]
            image_paths = image_paths[:10]

        prompt = "These are consecutive frames from a video. Describe this scene."
        images.append(prompt)

        response = model.generate_content(images)

        try:
          response_text = response.text
        except ValueError as ve:
          response_text = str(ve)
        except Exception as e:
          response_text = str(e)

        # Append the result to the output data
        output_data.append({
            "image_folder": folder_name,
            "image_paths" : image_paths,
            "response_text": response_text
        })

        print(folder_name)
        print(response_text)

# Create a DataFrame from the output data
geminiresponses = pd.DataFrame(output_data)
```
