# Re-compiling frames into video

```python
%cd /content/multi-hmr/video
import os
import cv2

if not os.path.exists('frames'):
    os.makedirs('frames')

video_path = '/content/multi-hmr/video/HMRtest1.mp4'

video = cv2.VideoCapture(video_path)

frame_count = 0

while True:
    ret, frame = video.read()

    if not ret:
        break

    frame_filename = f'/content/multi-hmr/video/frames/frame_{frame_count:04d}.jpg'

    cv2.imwrite(frame_filename, frame)

    frame_count += 1

video.release()

print(f'Total frames extracted: {frame_count}')

```

```python
import ffmpeg
import re
import os
from PIL import Image

original_video_path = '/content/drive/MyDrive/HMR/HMRtest1.mp4'
frames_folder_path = '/content/drive/MyDrive/HMR/multi-hmr_transparent/hmrtesttransparent'
output_video_path = 'transparentbg.mp4'

# Generate the input pattern for the frames
input_pattern = os.path.join(frames_folder_path, 'frame_%04d_multiHMR_896_L.png')

first_frame_path = os.path.join(frames_folder_path, 'frame_0000_multiHMR_896_L.png')
with Image.open(first_frame_path) as img:
    width, height = img.size
    print(width, height)


try:
    probe = ffmpeg.probe(original_video_path)
    video_info = next(stream for stream in probe['streams'] if stream['codec_type'] == 'video')
    framerate = eval(video_info['r_frame_rate'])

    ffmpeg_input = ffmpeg.input(input_pattern, framerate=framerate)
    ffmpeg_output = ffmpeg.output(ffmpeg_input, output_video_path, vcodec='libx264', pix_fmt='yuv420p', s='{}x{}'.format(width, height))

    ffmpeg.run(ffmpeg_output, capture_stdout=True, capture_stderr=True)
    print(f"Video created successfully: {output_video_path}")

except ffmpeg.Error as e:
    print(f"FFmpeg error: {e.stderr.decode('utf8')}")
```
