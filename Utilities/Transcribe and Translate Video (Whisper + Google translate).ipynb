# Transcribe batch of videos

```python
import re
import subprocess
transcriptions = []

# transcribe video files using Whisper
def transcribe_video(file_path):
    try:
        result = subprocess.run(
            ['whisper', file_path, '--model', 'large', '--language', 'Arabic'],
            capture_output=True,
            text=True
        )
        return result.stdout
    except Exception as e:
        print(f"An error occurred while transcribing {file_path}: {e}")
        return ""

# parse transcription output
def parse_transcription(transcription_text):
    pattern = re.compile(r'\[(\d{2}:\d{2}.\d{3})\s-->\s(\d{2}:\d{2}.\d{3})\]\s(.+)')
    matches = pattern.findall(transcription_text)
    parsed_results = []
    for match in matches:
        start_time, end_time, text = match
        parsed_results.append({
            'start_time': start_time,
            'end_time': end_time,
            'text': text
        })
    return parsed_results
```

```python
from tqdm.notebook import tqdm

source_dir = "/content/drive/MyDrive/Other/needtotranscribe"
results_dir = "/content/drive/MyDrive/Other/transcriptions"

if not os.path.exists(results_dir):
    os.makedirs(results_dir)

video_files = [f for f in os.listdir(source_dir) if any(f.lower().endswith(ext) for ext in video_extensions)]
for video_file in tqdm(video_files, desc="Transcribing videos"):
    video_path = os.path.basename(video_file)
    transcription_text = transcribe_video(video_path)
    parsed_transcriptions = parse_transcription(transcription_text)
    for entry in parsed_transcriptions:
        transcriptions.append([
            video_path,
            entry['start_time'],
            entry['end_time'],
            entry['text']
        ])

    checkpoint_file = os.path.join(results_dir, 'transcriptions_checkpoint.json')
    with open(checkpoint_file, 'w') as f:
        json.dump(transcriptions, f, ensure_ascii=False, indent=4)

transcriptions_np = np.array(transcriptions)

```
